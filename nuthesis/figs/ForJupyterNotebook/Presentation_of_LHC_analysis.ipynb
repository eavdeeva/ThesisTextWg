### 1.2. What is going on during the collision of two highly energetic protons

```python
Image(filename='Theory_ppCollision.png')
```

Let us take a closer look into the proton structure. We just discussed that protons consist of three quarks: $uud$. However, it is not the whole story. These three quarks interact with each other by exchanging gluons (curved lines), and these gluons produce quark-antiquark pairs that are called *sea* quarks and antiquarks. In contrast, the three original quarks are called *valence* quarks. All these proton constituents, valence quarks, sea quarks and antiquarks, and gluons, are called *partons*. 

When two protons smash, usually one parton from one of the protons interact with another parton from the other proton. It can be quark-quark ($qq$), quark-gluon ($qg$), quark-antiquark ($q\bar{q}$) collision etc. At the moment of a collision, these two partons disappear, and a new particles or a number of particles are born. At this particular figure, a $Z$ boson is born and further decay into $\mu^+ \mu^-$ pair (and each of these muons will decay microseconds later). In another collision it could be another primary particle, or a $Z$ boson again but decaying into other particles. 

The collision process is completely probabilistic. While certain collision scenarios are much more likely to happen than others, there is no way to predict what exactly is going to happen during the given collision.

The question is can we understand what was the process at least AFTER it already happened. This whole notebook will try to bring more insight to answer this question. I will start with explaining what particle collision process I'm looking for.
