### 1.1. Fundamental particles

There is an atom that consists of a nucleus and an electron orbiting around it. The electron is not known to have any structure. Thus, we call it a *fundamental particle*. At the same time, the nucleus consists of protons and neutrons. Each proton and neutron is a composite particle: they consist of quarks. Quarks are not known to have any structure. Thus, they are fundamental particles too.

### 1.2. What is going on during the collision of two highly energetic protons

```python
Image(filename='Theory_ppCollision.png')
```

Let us take a closer look into the proton structure. We just discussed that protons consist of three quarks: $uud$. However, it is not the whole story. These three quarks interact with each other by exchanging gluons (curved lines), and these gluons produce quark-antiquark pairs that are called *sea* quarks and antiquarks. In contrast, the three original quarks are called *valence* quarks. All these proton constituents, valence quarks, sea quarks and antiquarks, and gluons, are called *partons*. 

When two protons smash, usually one parton from one of the protons interact with another parton from the other proton. It can be quark-quark ($qq'$), quark-gluon ($qg$), quark-antiquark ($q\bar{q'}$) collision etc. At the moment of a collision, these two partons disappear, and a new particle or a number of particles are born. At this particular figure, a $W$ boson is born and further decay into a muon neutrino ($\nu_\mu$) and a muon ($\mu^+$) which later radiates a photon ($\gamma$). In the other collision, it could be another primary particle, or a $W$ boson again but decaying differently. 

The collision process is completely probabilistic. While certain collision scenarios are much more likely to happen than others, there is no way to predict what exactly is going to happen during the given collision.

The question is can we understand what was the process at least AFTER it already happened. This whole notebook will try to bring more insight to answer this question.
